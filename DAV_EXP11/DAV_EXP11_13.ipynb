*Categorical Variables*

```r
sns.countplot(x='Preferred Foot', data=df,stat="percent")
plt.title('Distribution of Preferred Foot')
plt.show()
```

```r
grouped_counts = df.groupby('Nationality').size().reset_index(name='Count')

# Sort by the count values
grouped_counts_sorted = grouped_counts.sort_values(by='Count', ascending=False)

print(grouped_counts_sorted.head(7))
```

```r
filtered_df = df[df['Nationality'].isin(grouped_counts_sorted.head(7)['Nationality'])]
```

```r
sns.countplot(x='Nationality', hue='Preferred Foot', data=filtered_df,stat="percent")
plt.title('Compairing the top 7 nationality prefered foot')
plt.show()
```

```r
sns.histplot(df['Height'], bins=10,kde=True,color='crimson', label='Height (in foot)')
plt.title('Height Distribution')
plt.show()
```

```r
sns.pairplot(df)
```

```r
library(dplyr)
library(ggplot2)
```

```r
df <- read.csv("/content/fifa_eda.csv")
```

```r
df
```

```r
options(repr.plot.width = 10, repr.plot.height = 6)
ggplot(df, aes(x = Height, y = Weight, color =Preferred.Foot)) +
  geom_point() +
  labs(title = "Scatter Plot of Height vs. Weight", x = "Height", y = "Weight") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```r
ggplot(df, aes(x = Age, fill =Preferred.Foot )) +
  geom_bar(position = "stack") +
  labs(title = "Stacked Bar Plot of Age by Preferred Foot", x = "Age", y = "Preferred Foot") +
   theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```r
ggplot(df, aes(x = Age)) +
  geom_line(aes(y = Overall,color = "Overall")) +
  geom_line(aes(y = Potential, color = "Potential")) +
  labs(title = "Line Plot of Overall and Potential stats",
       x = "Age",
       y = "") +
  theme_minimal() +
    theme(
    text = element_text(size = 16),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    plot.title = element_text(size = 20)
  )
```
